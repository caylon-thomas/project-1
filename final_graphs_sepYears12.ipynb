```python
# change over the years
change_10=df_10.groupby('Year')['Month'].count()
change_11=df_11.groupby('Year')['Month'].count()
change_12=df_12.groupby('Year')['Month'].count()
change_13=df_13.groupby('Year')['Month'].count()
change_14=df_14.groupby('Year')['Month'].count()
change_15=df_15.groupby('Year')['Month'].count()
change_16=df_16.groupby('Year')['Month'].count()
change_17=df_17.groupby('Year')['Month'].count()
change_18=df_18.groupby('Year')['Month'].count()

change_10=pd.DataFrame(change_10)
change_11=pd.DataFrame(change_11)
change_12=pd.DataFrame(change_12)
change_13=pd.DataFrame(change_13)
change_14=pd.DataFrame(change_14)
change_15=pd.DataFrame(change_15)
change_16=pd.DataFrame(change_16)
change_17=pd.DataFrame(change_17)
change_18=pd.DataFrame(change_18)

plt.figure(figsize=(10,8))
plt.bar(change_10.index, change_10['Month'], label='y-2010', align="center", alpha=0.5)
plt.bar(change_11.index, change_11['Month'], label='y-2011', align="center", alpha=0.5)
plt.bar(change_12.index, change_12['Month'], label='y-2012', align="center", alpha=0.5)
plt.bar(change_13.index, change_13['Month'], label='y-2013', align="center", alpha=0.5)
plt.bar(change_14.index, change_14['Month'], label='y-2014', align="center", alpha=0.5)
plt.bar(change_15.index, change_15['Month'], label='y-2015', align="center", alpha=0.5)
plt.bar(change_16.index, change_16['Month'], label='y-2016', align="center", alpha=0.5)
plt.bar(change_17.index, change_17['Month'], label='y-2017', align="center", alpha=0.5)
plt.bar(change_18.index, change_18['Month'], label='y-2018', align="center", alpha=0.5)

plt.xlim(2009.5,2018.5)
plt.legend(loc="upper left")
plt.xlabel("Years")
plt.ylabel("Visits")
plt.title('Visits by Year')
plt.tight_layout()
plt.savefig("years1.png")
plt.show()
```

```python
change_over_months_10=df_10.groupby('Month')['Date'].count()
change_over_months_11=df_11.groupby('Month')['Date'].count()
change_over_months_12=df_12.groupby('Month')['Date'].count()
change_over_months_13=df_13.groupby('Month')['Date'].count()
change_over_months_14=df_14.groupby('Month')['Date'].count()
change_over_months_15=df_15.groupby('Month')['Date'].count()
change_over_months_16=df_16.groupby('Month')['Date'].count()
change_over_months_17=df_17.groupby('Month')['Date'].count()
change_over_months_18=df_18.groupby('Month')['Date'].count()

plt.figure(figsize=(12,6))
plt.plot(change_over_months_10.index, change_over_months_10.values, label="y-2010", linewidth = 4, color='#0C1EA2')
plt.plot(change_over_months_11.index, change_over_months_11.values, label="y-2011", linewidth = 4, color='#136FA8')
plt.plot(change_over_months_12.index, change_over_months_12.values, label="y-2012", linewidth = 4, color='#3592CA')
plt.plot(change_over_months_13.index, change_over_months_13.values, label="y-2013", linewidth = 4, color='#91F7FD')
plt.plot(change_over_months_14.index, change_over_months_14.values, label="y-2014", linewidth = 4, color='#9BFD91')
plt.plot(change_over_months_15.index, change_over_months_15.values, label="y-2015", linewidth = 4, color='#26E47B')
plt.plot(change_over_months_16.index, change_over_months_16.values, label="y-2016", linewidth = 4, color='#10BF5F')
plt.plot(change_over_months_17.index, change_over_months_17.values, label="y-2017", linewidth = 4, color='#066C34')
plt.plot(change_over_months_18.index, change_over_months_18.values, label="y-2018", linewidth = 4, color='#053D20')

plt.xticks(np.arange(0, 13, 1))

plt.legend(loc="best")
plt.xlabel("Months")
plt.ylabel("Visits")
plt.title('Visits by Year over Months')

plt.tight_layout()
plt.savefig("months.png")

plt.show()
```
